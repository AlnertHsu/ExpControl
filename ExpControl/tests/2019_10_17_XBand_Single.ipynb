```python
plt.rcParams['lines.linewidth'] = 3.0
plt.rcParams['font.size'] = 22.0
plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
plt.rcParams['axes.labelsize'] = 26
plt.style.use('seaborn-white')
```

```python
meas = ExpControl.Measurements(instr=['vna'])
```

```python
meas.setup_vna(power=-30,avgs = 1,measure='S21')
```

```python
trace_wb_s11 = meas.trace(0e09,20.0e09,npts=20001,bandwidth=1000)
```

```python
trace_wb_s21 = meas.trace(0e09,20.0e09,npts=20001,bandwidth=1000)
```

```python
df_s11 = pd.DataFrame(data={'freq':trace_wb_s11[0],'ch1':trace_wb_s11[1][0], 'ch2':trace_wb_s11[1][1]})
```

```python
df_s21 = pd.DataFrame(data={'freq':trace_wb_s21[0],'ch1':trace_wb_s21[1][0], 'ch2':trace_wb_s21[1][1]})
```
