```python
import ExpControl
import numpy as np
import pandas as pd
import pickle
from matplotlib import pyplot as plt
```

## Setup the Instruments

```python
meas = ExpControl.Measurements(instr=['vna','prx'])
meas.setup_vna(power=-30,avgs = 1,measure = 'S21')
```

# Chip 3468

```python
trace_wb = meas.trace(2.0e09,12.0e09,npts=20001,bandwidth=1000)
```

```python
trace_nb = meas.trace(7.515e09,7.521e09,npts=5001,bandwidth=1000)
```

```python
da = ExpControl.DataAnalysis()
fit_3468 = da.LorentzianFit(trace_nb[0],trace_nb[1][0],plot=True) 
```

```python
df_3468_wb = pd.DataFrame(data = {'freq':trace_wb[0],'ch1':trace_wb[1][0],'ch2':trace_wb[1][1]})
df_3468_nb = pd.DataFrame(data = {'freq':trace_nb[0],'ch1':trace_nb[1][0],'ch2':trace_nb[1][1]})
df_3468_fit = pd.DataFrame(data = {'x':fit_3468.x, 'y':fit_3468.y, 'fit':fit_3468.fit, 
                'sigma':fit_3468.sigma, 'centre':fit_3468.centre, 'Q':fit_3468.Q})
df_3468_wb.to_pickle("data/3468_wb.pkl")
df_3468_nb.to_pickle("data/3468_nb.pkl")
df_3468_fit.to_pickle("data/3468_fit.pkl")
```
